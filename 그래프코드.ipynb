```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 60)
```

```python
fifa_final_data = pd.read_csv("C:\\Users\\박채은\\DSL\\피파온라인4\\final_data.csv")
fifa_final_data
```

```python
# seasonId가 numeric으로 되어 있어서 string 형태로 바꿔주기

fifa_final_data['seasonId']=fifa_final_data['seasonId'].apply(str)
```

```python
fifa_final_data.columns
```

```python
# matchEndType = 1,2 제거하기. 컴퓨터 튕겨서 나가진 경우
fifa_final_data = fifa_final_data[fifa_final_data["matchEndType"]==0]

# 패스시도=0 제거하기. 해당 시즌 넥슨 경기 기록 집계 오류로 인해 발생하는 문제점.
fifa_final_data = fifa_final_data[fifa_final_data["passTry"]!=0]

# 중복 데이터 제거하기
fifa_final_data = fifa_final_data.drop_duplicates()
fifa_final_data
```

```python
# 결측치 확인
fifa_final_data.isnull().sum()
```

```python
fifa_final_data = fifa_final_data.groupby(['seasonId'], as_index=False).mean()
fifa_final_data.head(3)
```

```python
# 데이터 성공률 계산

x = pd.DataFrame()
x["HeadingSuccessRate"]=(fifa_final_data["goalHeading"]/fifa_final_data["shootHeading"])
x["FreekickSuccessRate"]=(fifa_final_data["goalFreekick"]/fifa_final_data["shootFreekick"])
x["InPenalatySuccessRate"]=(fifa_final_data["goalInPenalty"]/fifa_final_data["shootInPenalty"])
x["OutPenaltySuccessRate"]=(fifa_final_data["goalOutPenalty"]/fifa_final_data["shootOutPenalty"])
x["PenaltyKickSuccessRate"]=(fifa_final_data["goalPenaltyKick"]/fifa_final_data["shootPenaltyKick"])
x["PassSuccessRate"]=(fifa_final_data["passSuccess"]/fifa_final_data["passTry"])
x["ShortPassSuccessRate"]=(fifa_final_data["shortPassSuccess"]/fifa_final_data["shortPassTry"])
x["LongPassSuccessRate"]=(fifa_final_data["longPassSuccess"]/fifa_final_data["longPassTry"])
x["BouncingLobPassSuccessRate"]=(fifa_final_data["bouncingLobPassSuccess"]/fifa_final_data["bouncingLobPassTry"])
x["DrivenGroundPassSuccessRate"]=(fifa_final_data["drivenGroundPassSuccess"]/fifa_final_data["drivenGroundPassTry"])
x["ThroughPassSuccessRate"]=(fifa_final_data["throughPassSuccess"]/fifa_final_data["throughPassTry"])
x["LobbedThroughPassSuccessRate"]=(fifa_final_data["lobbedThroughPassSuccess"]/fifa_final_data["lobbedThroughPassTry"])
x["BlockSuccessRate"]=(fifa_final_data["blockSuccess"]/fifa_final_data["blockTry"])
x["TackleSuccessRate"]=(fifa_final_data["tackleSuccess"]/fifa_final_data["tackleTry"])

fifa_final_data = pd.concat([fifa_final_data, x], axis=1)
fifa_final_data
```
