```python
for users in user_code : 
    
    match_ids = []
            
    for i in range(10): ## 여기가 매우 중요 ; 본인 월에 따라 호출량을 선택

        URL2 = "https://api.nexon.co.kr/fifaonline4/v1.0/users/{}/matches?matchtype={}&offset={}&limit={}".format(users, 50, 100*i, 100)
        res2 = req.get(URL2, headers=headers)

        # API 사용량 제한에 따라서 생기는 오류 처리
        if res2.status_code == 200:
            data = res2.json()
            for dat in data:
                if dat not in match_ids:
                    match_ids.append(dat)

        elif res2.status_code == 429:
            while res2.status_code != 200:
                print("API 사용량 한계 도달.. 10초 기다리는 중")
                time.sleep(10)
                res2 = req.get(URL2, headers=headers)

        else:
            print("알 수 없는 오류 발생.")
            
        ##### 이사람의 최근 n개의 매치번호가 출력됨 ##### ## match_ids에 할당
    print('User {}의 경기를 불러왔습니다!'.format(users))
    
    for match_id_ in match_ids : 

        URL3 = 'https://api.nexon.co.kr/fifaonline4/v1.0/matches/{}'.format(match_id_)
        res3 = req.get(URL3, headers=headers)

        if res3.status_code == 200 : 
            results.append(res3.json())

        elif res3.status_code == 429:
            while res3.status_code != 200:
                print("API 사용량 한계 도달.. 10초 기다리는 중")
                time.sleep(10)
                res3 = req.get(URL3, headers=headers)
                results.append(res3.json())
        else : 
            print('뭔가 이상하네용')
                
    print('User {} has been finished'.format(users))
```
